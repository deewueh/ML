2. **Столбчатая диаграмма для категориальной переменной `Class`:**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('Coffe_sales.csv')

categorical_cols = ["cash_type", "coffee_name"]

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

for ax, col in zip(axes.flatten(), categorical_cols):
    sns.countplot(x=col, data=df, ax=ax, order=df[col].value_counts().index)
    ax.set_title(f"Распределение по {col}")
    ax.set_xlabel('')
    ax.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

**1. Распределение по cash_type**
На этом графике всего один столбец — "card". Это означает, что 100% всех транзакций в этом наборе данных были совершены с помощью карты.
Переменная полностью сконцентрирована в одном значении. Это говорит о том, что данные либо собирались только по безналичным платежам, либо в этой кофейне в принципе не принимают наличные.

**2. Распределение по coffee_name**
График показывает популярность разных напитков. Столбцы отсортированы по убыванию количества продаж.
Наибольшее количество продаж сконцентрировано на двух напитках: "Americano with Milk" и "Latte". Они являются явными лидерами. Далее по популярности с небольшим отрывом идут "Americano" и "Cappuccino". Наименее популярным напитком является "Espresso".



**Боксплот для числовой переменной:**

```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
n_cols = len(num_cols)
n_rows = (n_cols + 2) // 3

fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(10, 5))
axes = axes.flatten()

for ax, col in zip(axes, num_cols):
   sns.boxplot(df[col], ax=ax)
   ax.set_title(f'Распределение {col}')
   ax.set_xlabel(col)
   ax.set_ylabel('Частота')

plt.tight_layout()
plt.show()
```

**1. Распределение hour_of_day**

Центральные 50% всех покупок совершаются между 10:00 и 18:00. Медианное время покупки — около 14:00. Весь диапазон покупок — с 6 утра до 22 вечера.
Выбросов нет, все значения находятся в пределах "усов", что говорит об отсутствии аномально ранних или поздних покупок.

**2. Распределение money**

50% всех чеков находятся в диапазоне примерно от 28 до 36. Медианная стоимость покупки — около 33.
Есть выбросы с обеих сторон. Есть несколько транзакций с аномально низкой стоимостью (ниже 20) и, возможно, с аномально высокой (хотя на графике верхних выбросов не видно, но верхний "ус" не доходит до максимального значения 38.7, что может указывать на них). Эти выбросы — это, вероятно, самые дешевые и самые дорогие напитки, которые покупают реже основной массы.

**3. Распределение Weekdaysort**

Этот график показывает распределение по дням недели от 1 (Понедельник) до 7 (Воскресенье). Медиана находится на отметке 4, что соответствует четвергу. 50% данных лежат в диапазоне от 2 (вторник) до 6 (суббота).
Выбросов нет, т.к. это порядковая переменная с ограниченным набором значений (1-7), выбросов здесь быть не может.

**4. Распределение Monthsort**

График показывает распределение по месяцам. Медиана находится на отметке 7 (июль), что означает, что половина данных приходится на первую половину года, а вторая — на вторую. Центральные 50% наблюдений приходятся на период с 3-го (март) по 10-й (октябрь) месяц.
Выыбросов нет, как и с днями недели, это порядковая переменная с фиксированным диапазоном (1-12), поэтому выбросов нет.

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Корреляционная матрица")
plt.show()
```

Переменные, у которых коэффициент корреляции близок к нулю (на графике они окрашены в синий или светло-голубой цвет), практически не имеют линейной связи. В данном случае это почти все пары:

hour_of_day и Weekdaysort (-0.0026): Связи нет. Время покупки не зависит от дня недели.
hour_of_day и Monthsort (0.0083): Связи нет. Время покупки не зависит от месяца.
money и Weekdaysort (-0.017): Связи нет. Сумма покупки не зависит от дня недели.
money и Monthsort (-0.05): Связи практически нет. Сумма покупки не зависит от месяца.
Weekdaysort и Monthsort (0.044): Связи нет. День недели и месяц никак не связаны (что логично).


Ни одна из пар различных переменных не имеет сильной положительной корреляции. Все значения, кроме единиц на диагонали, значительно ниже этого порога.

Единственная пара, которая демонстрирует какую-либо заметную связь, — это hour_of_day и money с коэффициентом 0.2.
Это очень слабая положительная корреляция. Она говорит о том, что есть небольшая, едва заметная тенденция: чем позже в течение дня совершается покупка, тем незначительно выше может быть ее стоимость. Однако эта связь слишком слаба, чтобы считать ее статистически или практически значимой.
